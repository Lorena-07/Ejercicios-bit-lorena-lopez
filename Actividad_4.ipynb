### 1. ¿Cual es el valor promedio de horas extras diurnas trabajadas por codigo de cargo?

```python
archivo.info()
```

```python
promedio_horas = archivo.groupby("CODIGO CARGO")["VALOR HORAS EXTRAS DIURNAS TRABAJADAS"].mean()
promedio_horas
```

### El valor promedio de horas extras trabajadas por codigo de cargo se discrima a continuacion.

```python
promedio_horas = promedio_horas.apply(lambda x: "${:.1f}k".format((x/1000)))
promedio_horas = promedio_horas.astype(str)
promedio_horas
```
