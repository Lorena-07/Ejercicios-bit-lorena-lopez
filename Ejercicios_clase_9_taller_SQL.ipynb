### 4. Lista ProductID, p.Name AS ProductName, c.Name AS CategoryName para productos con categoría
(INNER JOIN).

```python
consulta_inner_join = """
SELECT
    p.ProductID,
    p.Name AS ProductName,
    c.Name AS CategoryName
FROM
    SalesLT.Product AS p
INNER JOIN
    SalesLT.ProductCategory AS c ON p.ProductCategoryID = c.ProductCategoryID;
"""

respuesta_inner_join = pd.read_sql_query(consulta_inner_join, conn)
respuesta_inner_join
```

### 5. Obtén pedidos del año 2008 en adelante (o el año que aplique en tu base) con SalesOrderID, OrderDate, CustomerID, TotalDue y el nombre del cliente (FirstName + ' ' + LastName).
(JOIN + WHERE)

```python
consulta_pedidos_2008 = """
SELECT
    SOH.SalesOrderID,
    SOH.OrderDate,
    SOH.CustomerID,
    SOH.TotalDue,
    CONCAT(C.FirstName, ' ', C.LastName) AS CustomerName
FROM
    SalesLT.SalesOrderHeader AS SOH
INNER JOIN
    SalesLT.Customer AS C ON SOH.CustomerID = C.CustomerID
WHERE
    YEAR(SOH.OrderDate) >= 2008;
"""

respuesta_pedidos_2008 = pd.read_sql_query(consulta_pedidos_2008, conn)
respuesta_pedidos_2008
```

### 6. Para cada ProductID, suma la cantidad vendida (SUM(OrderQty)). Devuelve solo productos con ventas totales > 1000, ordenados de manera descendente.
Usa GROUP BY, HAVING y ORDER BY

```python
consulta_ventas_agrupadas = """
SELECT ProductID, SUM(OrderQty) Categoria FROM SalesLT.SalesOrderDetail GROUP BY ProductID HAVING SUM(OrderQty) > 10 ORDER BY Categoria DESC
"""

respuesta_ventas_agrupadas = pd.read_sql_query(consulta_ventas_agrupadas, conn)
respuesta_ventas_agrupadas
```

### 7. Productos sin ventas: lista productos que no aparecen en SalesLT.SalesOrderDetail.
Usa LEFT JOIN + WHERE

```python
consulta_productos_sin_ventas = """
SELECT
    P.ProductID,
    P.Name
FROM
    SalesLT.Product AS P
LEFT JOIN
    SalesLT.SalesOrderDetail AS SOD ON P.ProductID = SOD.ProductID
WHERE
    SOD.SalesOrderID IS NULL;
"""

respuesta_productos_sin_ventas = pd.read_sql_query(consulta_productos_sin_ventas, conn)
respuesta_productos_sin_ventas
```

### 8. Clientes por país: total de clientes por CountryRegion (usa Address + CustomerAddress).
Podrias necesitar: COUNT(DISTINCT CustomerID).

```python
consulta_clientes_por_pais = """
SELECT
    A.CountryRegion,
    COUNT(DISTINCT C.CustomerID) AS TotalCustomers
FROM
    SalesLT.Customer AS C
INNER JOIN
    SalesLT.CustomerAddress AS CA ON C.CustomerID = CA.CustomerID
INNER JOIN
    SalesLT.Address AS A ON CA.AddressID = A.AddressID
GROUP BY
    A.CountryRegion
ORDER BY
    TotalCustomers DESC;
"""

respuesta_clientes_por_pais = pd.read_sql_query(consulta_clientes_por_pais, conn)
respuesta_clientes_por_pais
```

### 9. Ticket promedio por año: AVG(TotalDue) por YEAR(OrderDate).

```python
consulta_ticket_promedio_anual = """
SELECT
    YEAR(OrderDate) AS OrderYear,
    MONTH(OrderDate) AS OrderMonth,
    DAY(OrderDate) AS OrderDay,
    AVG(TotalDue) AS AverageTicket
FROM
    SalesLT.SalesOrderHeader
GROUP BY
    YEAR(OrderDate),
    MONTH(OrderDate),
    DAY(OrderDate)
ORDER BY
    OrderYear;
"""

respuesta_ticket_promedio_anual = pd.read_sql_query(consulta_ticket_promedio_anual, conn)
respuesta_ticket_promedio_anual
```
